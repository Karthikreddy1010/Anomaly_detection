```python
# I removed this column because it has so many null values half of data so, I removed it from data
data=data.drop(columns=['plugin.battery.disconnected_at'])
```

```python
## lls.value.1 is divide by 10 to convert in to liters

data['lls.value.1']=data['lls.value.1']/10
```

```python
# Plotting the lls.value.1 to visualize any anomalies
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(data['timestamp'], data['lls.value.1'], marker='o')
plt.title('LLS Value Over Time')
plt.xlabel('Timestamp')
plt.ylabel('LLS Value')
plt.show()
```

```python
# Histogram of lls.value.1 to see the distribution
plt.figure(figsize=(10, 6))
plt.hist(data['lls.value.1'], bins=50)
plt.title('Distribution of LLS Value')
plt.xlabel('LLS Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np

# Calculate Z-scores
data['lls_zscore'] = (data['lls.value.1'] - data['lls.value.1'].mean()) / data['lls.value.1'].std()

# Identify outliers
outliers = data[np.abs(data['lls_zscore']) > 3]
print("Outliers:")
print(outliers)
```

### Z score= X - mean(lls.value.1)/ std(lls.value.1)
### For Example

### mean=130.31 std=43.43 X is a sample X=107.3

### Z= 107.3 - 130.31/ 43.43

## Z=-0.5299

```python

import seaborn as sns
from sklearn.ensemble import IsolationForest
# Feature Engineering: Create additional features
data['lls_diff'] = data['lls.value.1'].diff().fillna(0)
data['expected_fuel_consumption'] = data['position.speed'] * data['vehicle.mileage'] / 1000  # Replace with your formula
data['unrealistic_consumption'] = (data['lls.value.1'] > 200) | (data['lls.value.1'] < 0)
```

```python
# Select features for anomaly detection
features = ['lls.value.1', 'position.latitude', 'position.longitude', 'position.altitude', 'position.speed', 'vehicle.mileage', 'lls_diff', 'expected_fuel_consumption']

# Fit the Isolation Forest model
model = IsolationForest(contamination=0.01, random_state=42)
data['anomaly'] = model.fit_predict(data[features])

# Add Target column: 0 for anomaly, 1 for non-anomaly
data['Target'] = data['anomaly'].apply(lambda x: 1 if x == 1 else 0)
```

```python
# Identify anomalies
anomalies = data[data['Target'] == 0]
print("Anomalies detected by Isolation Forest:")
print(anomalies)

# Plotting the anomalies
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data, x='position.speed', y='lls.value.1', hue='Target', palette={1: 'blue', 0: 'red'})
plt.title('LLS Value vs. Speed (Anomalies Highlighted)')
plt.xlabel('Position Speed')
plt.ylabel('LLS Value')
plt.legend(title='Target', loc='upper left', labels=['Non-Anomaly', 'Anomaly'])
plt.show()
```

```python
# Further Analysis of Anomalies
print("Detailed anomaly analysis:")
print(anomalies[['timestamp', 'position.latitude', 'position.longitude', 'position.altitude', 'position.speed', 'lls.value.1', 'vehicle.mileage', 'lls_diff', 'expected_fuel_consumption', 'unrealistic_consumption']])
```
